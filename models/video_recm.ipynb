```python
df['title'] = df['title'].fillna('')
df['description'] = df['description'].fillna('')
df['transcript'] = df['transcript'].fillna('')
```

```python
import pickle
df['combined'] = df['title'] + " " + df['description'] + " " + df['transcript']
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(df['combined'])

# def generate_recommendations(query, top_n=5):
#     query_tfidf = tfidf_vectorizer.transform([query])
#     cosine_similarities = cosine_similarity(query_tfidf, tfidf_matrix).flatten()
#     top_n_indices = cosine_similarities.argsort()[-top_n:][::-1]
#     recommended_videos = df.iloc[top_n_indices][['_id','title','thumbnail']]
#     return recommended_videos


# Define an object to be pickled (includes the function and necessary data)
class RecommendationModel:
    def __init__(self, vectorizer, matrix, data):
        self.vectorizer = vectorizer
        self.matrix = matrix
        self.data = data
    
    # The recommendation function needs access to the vectorizer and matrix
    def generate_recommendations(self, query, top_n=5):
        query_tfidf = self.vectorizer.transform([query])
        cosine_similarities = cosine_similarity(query_tfidf, self.matrix).flatten()
        top_n_indices = cosine_similarities.argsort()[-top_n:][::-1]
        return self.data.iloc[top_n_indices][['title', 'description', 'transcript']]

# Create an instance of the model with vectorizer, matrix, and data
recommendation_model = RecommendationModel(tfidf_vectorizer, tfidf_matrix, df)

# Serialize (pickle) the recommendation model to a file

```

```python
with open('recommendation_model.pkl', 'wb') as file:
    pickle.dump(recommendation_model, file)
```
